### Soal 3 (30 poin)
Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.
_Trend analysis_ dilakukan dengan menggunakan hubungan berikut:
![output%2011.png](https://www.dl.dropboxusercontent.com/s/dxmrm74ksz5l9v2/trend_analysis_fig.png?dl=0)

#### Code

```python
import datetime
```

```python
# type your code here
hang_seng = pd.read_csv('drive/My Drive/visdat/Hang_Seng.csv')
nasdaq = pd.read_csv('drive/My Drive/visdat/Nasdaq.csv')
nikkei = pd.read_csv('drive/My Drive/visdat/Nikkei.csv')
sample = pd.read_csv('drive/My Drive/visdat/Nikkei.csv')
```

```python
start_date = datetime.date(2019, 1 , 1)
number_of_days = 365

date_list = [(start_date + datetime.timedelta(days = day)).isoformat() for day in range(number_of_days)]
date_list[1]
```

```python
#Hang Seng 2019
hang_seng19 = sample
hang_seng19.drop(hang_seng19.index, inplace=True)
for i in range(365):
  hang_seng19 = pd.concat([hang_seng19,hang_seng.loc[hang_seng['Date'] == date_list[i]]])

hang_seng19["Day_Perc_Change"] = hang_seng19["Adj Close"].pct_change()*100
hang_seng19
```

```python
#Nasdaq 2019
nasdaq19 = sample
nasdaq19.drop(nasdaq19.index, inplace=True)
for i in range(365):
  nasdaq19 = pd.concat([nasdaq19,nasdaq.loc[nasdaq['Date'] == date_list[i]]])

nasdaq19["Day_Perc_Change"] = nasdaq19["Adj Close"].pct_change()*100
nasdaq19
```

```python
#Nikkei 2019
nikkei19 = sample
nikkei19.drop(nikkei19.index, inplace=True)
for i in range(365):
  nikkei19 = pd.concat([nikkei19,nikkei.loc[nikkei['Date'] == date_list[i]]])

nikkei19["Day_Perc_Change"] = nikkei19["Adj Close"].pct_change()*100
nikkei19
```

```python
# type your code here
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"
    
# hang_seng
hang_seng19["Trend"]= hang_seng19["Day_Perc_Change"].apply(lambda x:trend(x))
hang_seng19.dropna(axis = 0, inplace = True)
hang_seng19.head()
```

```python
hang_seng19.info()
```

```python
hangseng_pie_data = hang_seng19.groupby('Trend')
pie_label = sorted([i for i in hang_seng19.loc[:, 'Trend'].unique()])
plt.pie(hangseng_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)

plt.show()
```

Penjelasan

Untuk periode 2019, saham dari Heng Seng yang berada di antara pemenang teratas ada 0.4% saat itu. Sekitar 13.1% saat itu, saham Heng Seng menunjukan status positive di 2019. Demikian juga, lumayan banyak waktu (sekitar 26.1%) yang menunjukkan sedikit perubahan pada harga saham.

```python
# Nasdaq
nasdaq19["Trend"]= nasdaq19["Day_Perc_Change"].apply(lambda x:trend(x))
nasdaq19.dropna(axis = 0, inplace = True)
nasdaq19.head()
```

```python
nasdaq19.info()
```

```python
nasdaq_pie_data = nasdaq19.groupby('Trend')
pie_label = sorted([i for i in nasdaq19.loc[:, 'Trend'].unique()])
plt.pie(nasdaq_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)

plt.show()
```

Penjelasan

Untuk periode 2019, saham dari Nasdaq yang berada di antara pemenang teratas ada 0.4% saat itu, dan berada di antara top losers sebanyak 1.6%. Sebanyak 17.1% saat itu, saham Nasdaq menunjukan status positive di 2019. Sekitar 25.9% dari 2019 yang menunjukkan sedikit perubahan pada harga saham.

```python
# Nikkei
nikkei19["Trend"]= nikkei19["Day_Perc_Change"].apply(lambda x:trend(x))
nikkei19.dropna(axis = 0, inplace = True)
nikkei19.head()
```

```python
nikkei19.info()
```

```python
nikkei_pie_data = nikkei19.groupby('Trend')
pie_label = sorted([i for i in nikkei19.loc[:, 'Trend'].unique()])
plt.pie(nikkei_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)

plt.show()
```

Penjelasan

Untuk periode 2019, saham dari Nasdaq yang berada di antara pemenang teratas selama 0.4% dari 2019. Sebanyak 11.7% dari 2019, saham Nasdaq menunjukan status positive. Sekitar 27.9% dari 2019 menunjukkan sedikit perubahan pada harga saham.

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
# type your code here
combined_df = pd.concat([nasdaq19["Day_Perc_Change"], nikkei19["Day_Perc_Change"], hang_seng19["Day_Perc_Change"]], axis=1) # merge daily percentage change values
combined_df.columns = ["Nasdaq", "Nikkei", "Hang Seng"] # rename column
combined_df.dropna(inplace=True) # drop null values
combined_df = combined_df.reset_index() # reset index biar mulai dari 0
combined_df = combined_df.drop(columns=['index']) # drop column index
combined_df.head()
```

```python
combined_df.info()
```

```python
volatility = combined_df[['Nasdaq', 'Nikkei', 'Hang Seng']].rolling(7).std()*np.sqrt(7)
volatility.plot(figsize = (15, 7))
```

#### Penjelasan 
 
Volatile dari ketiga saham ini hampir sama. Pada awal tahun saham 'Nikkei' volatilenya cukup tinggi dibandingkan saham yang lainnya. Kemudian ketika volatile saham 'Nikkei' turun, saham yang lainnya mulai naik. Kemudian saham 'Nasdaq' volatilenya meningkat tinggi. Kemudian di akhir, saham 'Hang Seng' volatilenya meningkat cukup tinggi dibanding saham yang lain.
