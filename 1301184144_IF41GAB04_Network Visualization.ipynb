```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
```

```python
G.add_edges_from([('BANDUNG', 'CIMAHI'), ('BANDUNG', 'SOREANG'), ('BANDUNG', 'SUMEDANG'), ('BANDUNG', 'GARUT'), 
                  ('BANDUNG', 'TASIKMALAYA'),('GARUT', 'TASIKMALAYA'), ('BANDUNG', 'SUBANG'), ('CIMAHI', 'SOREANG'), ('CIMAHI', 'CIANJUR'), ('CIMAHI', 'PURWAKARTA')
                  , ('PURWAKARTA', 'SUBANG'), ('CIANJUR', 'SUKABUMI'), ('CIANJUR', 'BOGOR'), ('BOGOR', 'CIBINONG'), ('CIBINONG', 'DEPOK')
                  , ('DEPOK', 'BEKASI'), ('BEKASI', 'KARAWANG'), ('KARAWANG', 'PURWAKARTA'), ('TASIKMALAYA', 'CIAMIS')
                  , ('CIAMIS', 'KUNINGAN'), ('KUNINGAN', 'SUMBER'), ('SUMBER', 'CIREBON'), ('CIREBON', 'INDRAMAYU'), ('SUMEDANG', 'MAJALENGKA')
                  , ('MAJALENGKA', 'KUNINGAN'), ('MAJALENGKA', 'INDRAMAYU'), ('KARAWANG', 'INDRAMAYU')])
```

```python
fig, ax = plt.subplots(figsize=(25, 15))
nx.draw(G, with_labels = True)
plt.show()
```
